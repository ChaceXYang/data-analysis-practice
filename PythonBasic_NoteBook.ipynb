### input()函数
```

```
- prompt:如果指定该参数，那么将在标准输出中将参数的内容显示出来（末尾不会自动追加换行）

```

```

## 比较运算符
单独一个等于号（=）表示赋值运算符，作用是将右边的值跟左边的变量名进行挂钩。

|   运算符  | 含义               |
| :----: | ---------------- |
|    <   | 判断左边是否小于右边       |
|   <=   | 判断左边是否小于或等于右边    |

## 改进我们的小游戏

````python
"""用python设计第一个游戏"""
import random

answer = random.randint(1,10)
counts=3
while counts>0:
  temp=input("不妨猜一下小甲鱼现在心里面想的是哪个数字")
  guess=int(temp)
  if guess==answer:
    print("你是小甲鱼心里的蛔虫嘛")
    print("哼，猜中了也没有奖励")
    break
  else:
    if guess < answer:
      print("小啦~")
    else:
      print("大啦~")
  counts=counts-1
print("游戏结束，不玩啦^_^")
````

## 数字类型
Python 有三种不同的数字类型，分别是：整数、浮点数和复数。
### 1.整数
Python 的整数长度是不受限制的，也就是说它是有无限大的精度。
所以，你可以随时随地的进行大数运算：

````python
print(1112234579121621 / 11235813213455)
````

在Python中，整数的精度是**无限大**。  
这是因为Python的整数类型（`int`）采用动态内存分配机制，可以自动扩展以存储任意大小的整数（仅受计算机内存限制）

### 2.浮点数
我们通常数学意义上的小数在编程里叫浮点数。
- python和C一样使用IEEE754标准存储浮点数，会产生精度上的误差。

由于浮点数在计算机中的存储是存在 “误差” 的，所以有时候可能会闹出一些 “BUG”：

````python
print(0.1 + 0.2)
i=0
while i<1:
  i=i+0.1
  print(i)
````

由于浮点数并不是 100% 精确的，所以我们拿浮点数来做比较就要特别小心了：


````python
print(0.3==0.1+0.2)
````

0.1+0.2在python中存储的值可见开头，实际上是0.30000000000000004而不是0.3

#### 精确计算浮点数
这里我们需要借助到一个叫做 decimal 的模块
### decimal -- 十进制定点和浮点运算
decimal 模块支持快速正确舍入的十进制浮点运算。

````python
import decimal
a=decimal.Decimal('0.1')
b=decimal.Decimal('0.2')
print(a+b)
c=decimal.Decimal('0.3')
print(a+b==c)
print(0.00005)#python选择使用科学计数法 表述
````

它相对于 float 数据类型提供了几个优点：

- 十进制“是基于一个浮点模型，它是为人们设计的，并且必须有一个最重要的指导原则 - 计算机必须提供一个算法，与人们在学校学习的算法一样。 - 摘自十进制算术规范。

- 十进制数可以精确表示。
  
  相反，诸如 1.1 和 2.2 的数字在二进制浮点中没有精确表示。
  最终用户通常不会期望 1.1 + 2.2 显示为 3.3000000000000003，因为它与二进制浮点一样。

- 精确性转化为算术。
  
  在十进制浮点中，0.1 + 0.1 + 0.1 - 0.3 正好等于零。
  在二进制浮点中，结果为 5.5511151231257827e-017。
  
  当接近零时，差异阻止可靠的相等测试，并且差异可累积。
  
  因此，在具有严格相等不变量的会计应用程序中，十进制是首选。

- 十进制模块包含重要位置的概念，因此 1.30 + 1.20 是 2.50 。
  
  尾随零被保留以指示重要性。这是货币应用的习惯表述。
  
  对于乘法，“教科书”方法使用被乘数中的所有数字。
  例如，1.3 * 1.2 给出 1.56，而 1.30 * 1.20 给出 1.5600 。

### 3.复数

## 数字运算
Python 支持的数字运算如下：

| 操作 | 结果 |
| :--- | :--- |
| `x + y` | x 加 y 的结果 |
| `x - y` | x 减 y 的结果 |
| `x * y` | x 乘以 y 的结果 |
| `x / y` | x 除以 y 的结果 |
| `x // y` | x 除以 y 的结果（地板除） |
| `x % y` | x 除以 y 的余数 |
| `-x` | x 的相反数 |
| `+x` | x 本身 |
| `abs(x)` | x 的绝对值 |
| `int(x)` | 将 x 转换成整数 |
| `float(x)` | 将 x 转换成浮点数 |
| `complex(re, im)` | 返回一个复数，re 是实部，im 是虚部 |
| `c.conjugate()` | 返回 c 的共轭复数 |
| `divmod(x, y)` | 返回 `(x // y, x % y)` |
| `pow(x, y)` | 计算 x 的 y 次方 |
| `x ** y` | 计算 x 的 y 次方 |

- **地板除：**
  - 双斜杠（//）表示一种特殊的除法 —— 地板除。
  - 板除原理是**取比目标结果小的最大整数**。
  - 地板除的结果乘以除数 + 余数 = 被除数：`x == (x // y) * y + (x % y)`
- **取余：** 百分号（%）用于求两数相除的余数，如果能够整除，则余数为 0
- **divmod()函数：** Python 有个内置函数叫 divmod()，它的作用就是同时求出两参数地板除的结果和余数
- **abs()函数：** abs()函数的作用是返回指定数值的绝对值。如果传入的是一个复数，abs()函数返回的结果就是复数的模
- **int()，float() 和 complex() 函数**：
  - **int()函数**是将指定的值转换成整数，比如我们传入一个字符串'520'，那么得到结果就是一个整数520。不过如果参数是一个浮点数，那么就要注意了，因为它得到的将是一个截掉小数的整数。**注意：它是直接截取整数部分，扔掉小数部分，而不是四舍五入。**
  - 同样的道理，float() 和 complex() 函数是将指定的值转换成浮点数和复数
- **pow()函数**和幂运算符（**）：
  - 通常情况下，pow() 函数和幂运算符（**）这两个实现的效果是等价的
  - 不过，pow() 函数还留有一手，它支持第 3 个参数。如果传入第 3 个参数，那么会将幂运算的结果和第 3 个参数进行取余数运算

## 分支和循环branch and loop
### 1.分支结构

Python 的分支结构由 if 语句来操刀实现。

Python是通过缩进来检测代码块的，向右缩进相同距离的所有语句，都同属于一个代码块。

if 语句总共有 5 钟语法结构，其中前 4 种是比较常见的，而最后 1 种是比较炫酷的操作。


- **第1种**是判断一个条件，如果这个条件成立，就执行其包含的某条语句或某个代码块。

  语法结构如下：
  ```
  if condition条件:
      statement(s)某条语句或某个代码块
  ```
- **第2种**同样是判断一个条件，跟第 1 种的区别是如果条件不成立，则执行另外的某条语句或某个代码块。

  语法结构如下：
  ```
  if 条件:
    某条语句或某个代码块
  else:
    某条语句或某个代码块
  ```
- **第3种**是判断多个条件，如果第1个条件不成立，则继续判断第2个条件，如果第2个条件还不成立，则接着判断第3个条件……

  如果还有第 4、5、6、7、8、9 个条件，你还可以继续写下去。

  语法结构如下：
  ```
  if 第1个条件:
      某条语句或某个代码块
  elif 第2个条件:
      某条语句或某个代码块
  elif 第3个条件：
      某条语句或某个代码块
  ```
- **第4种**是在第3种的情况下添加一个else，表示上面所有的条件均不成立的情况下，执行某条语句或某个代码块。
  
  语法结构如下：
  ```
  if 第1个条件:
      某条语句或某个代码块
  elif 第2个条件:
      某条语句或某个代码块
  elif 第3个条件：
      某条语句或某个代码块
  else:
      某条语句或某个代码块
  ```
- **条件表达式**：它其实是一个条件表达式，相当于将一个完整的 if-else 结构整合成一个表达式来使用。
  ```
  条件成立时执行的语句 if 条件 else 条件不成立时执行的语句
  ```
  它把条件放正中间，然后左右紧挨着关键字 if 和 else，最左侧是条件成立时执行的语句，最右侧是条件不成立时执行的语句。

#### 3.continue 语句
实现跳出循环体还有另外一个语句，那就是 continue 语句。

continue 语句也会跳出循环体，但是，它只是跳出本一轮循环，它还会回到循环体的条件判断位置，然后继续下一轮循环（如果条件还满足的话）。

注意它和 break 语句两者的区别：
- continue 语句是跳出本次循环，回到循环的开头
- break 语句则是直接跳出循环体，继续执行后面的语句

# Python Tutorial for Beginners 4: Lists, Tuples, and Sets

## 一、Lists（列表）

### 1.1 基本概念与创建
列表是**有序的、可变的**序列数据类型，使用方括号 `[]` 创建，元素之间用逗号分隔。

```python
courses = ['history', 'math', 'physics', 'comp sci']
```

### 1.2 基础操作

#### 获取列表长度
使用 `len()` 函数返回列表元素个数：
```python
len(courses)  # 返回 4
```

#### 索引访问（Indexing）
- **正向索引**：从 `0` 开始，第一个元素索引为 0
- **反向索引**：从 `-1` 开始，最后一个元素索引为 -1
- **越界错误**：访问不存在的索引会抛出 `IndexError: list index out of range`

```python
courses[0]    # 'history'（第一个元素）
courses[3]    # 'comp sci'（第四个元素）
courses[-1]   # 'comp sci'（最后一个元素，推荐使用）
courses[4]    # IndexError：列表只有4个元素，最大索引为3
```

### 1.3 切片（Slicing）
切片用于获取列表的一个子集，语法为 `[start:stop]`，**包含 start，不包含 stop**。

```python
# 获取前两个元素（索引 0 和 1）
courses[0:2]   # ['history', 'math']
courses[:2]    # 同上，起始索引可省略，默认为0

# 获取后两个元素（索引 2 和 3）
courses[2:]    # ['physics', 'comp sci']，结束索引可省略，默认为末尾

# 完整切片
courses[:]     # 返回列表的浅拷贝
```

**注意**：切片不会修改原列表，返回新列表。

### 1.4 修改列表的方法（Mutable Methods）

#### 添加元素

| 方法 | 用法 | 说明 |
|------|------|------|
| `append()` | `courses.append('art')` | 在**末尾**添加单个元素 |
| `insert()` | `courses.insert(0, 'art')` | 在指定**索引位置**插入元素，原元素后移 |
| `extend()` | `courses.extend(courses_2)` | 将另一个**可迭代对象**的元素逐个添加到列表末尾 |

**重要区别**：
```python
courses_2 = ['art', 'education']

# 错误示范：使用 insert 或 append 添加列表
courses.insert(0, courses_2)  
# 结果：[['art', 'education'], 'history', 'math', 'physics', 'comp sci']
# 整个列表被当作单个元素插入，形成嵌套列表

# 正确示范：使用 extend
courses.extend(courses_2)
# 结果：['history', 'math', 'physics', 'comp sci', 'art', 'education']
# 列表被"展开"，元素逐个添加
```

#### 删除元素

| 方法 | 用法 | 说明 |
|------|------|------|
| `remove()` | `courses.remove('math')` | 根据**值**删除第一个匹配项，不存在则报错 |
| `pop()` | `courses.pop()` | 删除并**返回**指定索引的元素（默认最后一个），常用于栈/队列操作 |

```python
# pop() 的使用示例
popped_item = courses.pop()      # 删除并返回最后一个元素
popped_item = courses.pop(0)     # 删除并返回第一个元素
```

#### 排序与反转

| 方法 | 用法 | 说明 |
|------|------|------|
| `reverse()` | `courses.reverse()` | **原地**反转列表顺序，无返回值 |
| `sort()` | `courses.sort()` | **原地**排序（数字升序/字母字母序），无返回值 |

```python
# 降序排序
courses.sort(reverse=True)

# 数字列表排序
nums = [3, 1, 4, 1, 5]
nums.sort()          # [1, 1, 3, 4, 5]（升序）
nums.sort(reverse=True)  # [5, 4, 3, 1, 1]（降序）
```

**重要提示**：`reverse()` 和 `sort()` 都是**原地操作**（in-place），直接修改原列表，返回 `None`。

### 1.5 不改变原列表的函数

#### sorted() 函数
返回排序后的**新列表**，原列表保持不变。

```python
sorted_courses = sorted(courses)  # courses 保持不变
```

#### 内置数学函数
适用于数字列表：
```python
nums = [1, 5, 2, 4, 3]

min(nums)    # 1（最小值）
max(nums)    # 5（最大值）
sum(nums)    # 15（求和）
```

### 1.6 查找与成员检查

#### index() 方法
返回指定值的**第一个匹配索引**，不存在则抛出 `ValueError`。

```python
courses.index('comp sci')   # 返回 3
courses.index('art')        # ValueError: 'art' is not in list
```

#### in 运算符（成员检查）
返回布尔值 `True` 或 `False`，用于条件判断。

```python
'math' in courses      # True
'art' in courses       # False
```

### 1.7 遍历列表

#### 基础 for 循环
```python
for item in courses:
    print(item)
```

#### 使用 enumerate() 获取索引和值
`enumerate()` 返回索引和值的元组，可指定起始编号。

```python
# 默认从 0 开始
for index, course in enumerate(courses):
    print(index, course)

# 指定从 1 开始（常用于显示排名、序号）
for index, course in enumerate(courses, start=1):
    print(index, course)  # 1 history, 2 math...
```

### 1.8 列表与字符串的转换

#### join() - 列表转字符串
**字符串方法**，用指定分隔符连接列表元素（列表元素必须是字符串）。

```python
courses = ['history', 'math', 'physics', 'comp sci']

# 使用逗号加空格连接
course_str = ', '.join(courses)
# 结果："history, math, physics, comp sci"

# 使用连字符连接
course_str = ' - '.join(courses)
# 结果："history - math - physics - comp sci"
```

#### split() - 字符串转列表
**字符串方法**，按指定分隔符分割字符串为列表。

```python
# 将上面的字符串转回列表
new_list = course_str.split(' - ')
# 结果：['history', 'math', 'physics', 'comp sci']
```

## 二、Tuples（元组）

### 2.1 基本概念
元组与列表非常相似，但有两个关键区别：
- 使用**圆括号** `()` 定义（也可省略括号）
- **不可变（Immutable）**：创建后不能修改、添加或删除元素

```python
tuple_1 = ('history', 'math', 'physics', 'comp sci')
```

### 2.2 不可变性的重要性

#### 可变对象的陷阱（以列表为例）
当两个变量指向同一个列表时，修改其中一个会影响另一个：

```python
list_1 = ['history', 'math', 'physics', 'comp sci']
list_2 = list_1

list_1[0] = 'art'
print(list_2[0])  # 'art' - list_2 也被修改了！
```

#### 元组的保护作用
元组的不可变性使其成为**数据保护**的理想选择：

```python
tuple_1 = ('history', 'math', 'physics', 'comp sci')
tuple_2 = tuple_1

# 尝试修改会引发 TypeError
tuple_1[0] = 'art'  # TypeError: 'tuple' object does not support item assignment
```

### 2.3 元组的方法与操作
由于不可变性，元组只有以下方法：
- `count(x)`：统计元素 x 出现的次数
- `index(x)`：返回元素 x 的索引

**支持的操作**：索引访问、切片、遍历、`in` 运算符、`len()` 函数等（与列表相同）。

### 2.4 何时使用元组？
- 数据不应被意外修改（如常量配置、函数多返回值）
- 作为字典的键（列表不能作为键，因为不可哈希）
- 提高代码安全性，防止副作用

## 三、Sets（集合）

### 3.1 基本概念
集合是**无序的、不重复的**元素集合，使用**花括号** `{}` 创建。

关键特性：
- **无序性**：不记录元素添加顺序，每次打印顺序可能不同
- **唯一性**：自动去除重复元素
- **高效性**：基于哈希表实现，成员检查比列表快得多

```python
cs_courses = {'history', 'math', 'physics', 'comp sci', 'math'}  
# 结果只有4个元素，重复的 'math' 被自动去除
```

### 3.2 主要用途

#### 1. 去重
快速去除序列中的重复值：
```python
nums = [1, 1, 2, 2, 3, 3]
unique_nums = set(nums)  # {1, 2, 3}
```

#### 2. 成员检查（Membership Test）
检查元素是否存在于集合中，时间复杂度为 O(1)，远快于列表的 O(n)：

```python
'math' in cs_courses  # True（效率极高）
```

#### 3. 集合运算
假设有两个集合：
```python
cs_courses = {'history', 'math', 'physics', 'comp sci'}
art_courses = {'history', 'math', 'art', 'design'}
```

| 方法 | 运算符 | 结果 | 说明 |
|------|--------|------|------|
| `intersection()` | `&` | `{'history', 'math'}` | 交集：两个集合共有的元素 |
| `difference()` | `-` | `{'physics', 'comp sci'}` | 差集：在 CS 中但不在 Art 中的元素 |
| `union()` | `|` | `{'history', 'math', 'physics', 'comp sci', 'art', 'design'}` | 并集：两个集合所有的唯一元素 |
| `symmetric_difference()` | `^` | `{'physics', 'comp sci', 'art', 'design'}` | 对称差集：仅在其中一个集合中的元素 |

代码示例：
```python
# 交集
cs_courses.intersection(art_courses)  # {'history', 'math'}

# 差集（CS 有但 Art 没有）
cs_courses.difference(art_courses)    # {'physics', 'comp sci'}

# 并集（所有课程）
cs_courses.union(art_courses)         # 6个元素
```

### 3.3 空数据结构的创建

**特别注意**：创建空集合时不能 `{}`，因为 `{}` 创建的是空字典！

```python
# 空列表
empty_list = []
empty_list = list()

# 空元组
empty_tuple = ()
empty_tuple = tuple()

# 空集合（正确方式）
empty_set = set()

# ⚠️ 错误！这创建的是空字典，不是集合
wrong_empty = {}  # 这是一个 dict
```

1. **切片原理**：`[start:stop]` 包含 start，不包含 stop，这是 Python 的惯例设计
